```python
#Getting todays weather from OpenweatherMap website using Api and putting it in dataframe 
import requests
api_key='e2ae56b81f6d964b3945207092269084'
result_set=[]
city_list=df['City'].values.tolist()
for city in city_list:
    temp=[]
    api_adress='http://api.openweathermap.org/data/2.5/weather?appid='+api_key+'&q='+city
    data=requests.get(api_adress).json()
    temp.append(city)
    temp.append(data['weather'][0]['main'])
    temp.append(round(data['main']['temp']-273))
    temp.append(data['visibility']/1000)
    temp.append(data['wind']['speed'])
    temp.append(data['main']['pressure'])
    result_set.append(temp)
today_weather=pd.DataFrame(result_set,columns = ['City','Today weather', "Temperature in Celcius",'Visibility in Km','Wind Speed m/s','Pressure in hPa'])
today_weather.head()
```

```python
#Mearging created dataframe and extracted dataframe on City column
final=pd.merge(df,today_weather,on=['City'])
final.head()
```
